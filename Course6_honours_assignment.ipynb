```python
## EX1.Q2 :Write and execute a SQL query to list all crimes that took place at a school. Include case number, crime type and community name
%sql select p.case_number, p.primary_type, p.location_description, c.community_area_name from chicago_crime_data p inner join census_data c on p.community_area_number = c.community_area_number where location_description = 'SCHOOL, PUBLIC, GROUNDS'
```
